**Iris** **Flower** **Classification**

```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
sns.set(style="white",color_codes=True)

import warnings
warnings.filterwarnings('ignore')
```

```python
iris=pd.read_csv("/content/Iris.csv")
```

```python
iris.head()
```

```python
iris["Species"].value_counts()
```

```python
sns.FacetGrid(iris, hue="Species",height=6).map(plt.scatter,"PetalLengthCm","SepalWidthCm").add_legend()
```

```python
flower_map={'Iris-setosa':0,'Iris-versicolor':1,'Iris-virginica':2}
```

```python
iris["Species"]=iris["Species"].map(flower_map)
```

```python
iris.head()
```

```python
x=iris[['SepalLengthCm','SepalWidthCm','PetalLengthCm','PetalWidthCm']].values
```

```python
y=iris[['Species']].values
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression()
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
expected=y
predicted=model.predict(x)
predicted
```

```python
from sklearn import metrics
```

```python
print(metrics.classification_report(expected,predicted))
```

```python
print(metrics.confusion_matrix(expected,predicted))
```

```python
model=LogisticRegression(C=20,penalty='l2')
```

```python
model.fit(x,y)
```

```python
model.score(x,y)
```

```python
input_data=(5.1,3.5,1.4,0.2)
```

```python
import numpy as np
```

```python
input_numpy=np.asarray(input_data)
```

```python
input_reshape=input_numpy.reshape(1,-1)
```

```python
prediction=model.predict(input_reshape)
```

```python
print(prediction)
```

```python
if(prediction==1):
  print('Iris-versicolor')
elif (prediction==2):
  print('Iris-virginica')
else:
  print('Iris-setosa')
```
